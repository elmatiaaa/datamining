# Tugas 8 (Ensembel Learning)

Ensemble learning adalah paradigma pembelajaran mesin di mana banyak pelajar dilatih untuk memecahkan masalah yang sama. Berbeda dengan pendekatan pembelajaran mesin biasa yang mencoba mempelajari satu hipotesis dari data pelatihan, metode ansambel mencoba membangun serangkaian hipotesis dan menggabungkannya untuk digunakan.

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/08-Ahlaqul-Karimah/machine-Learning/main/mnist_test.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
xtr = df.iloc[:,1:]
ytr = df.iloc[:,0]

xtst = df.iloc[:,1:]
ytst = df.iloc[:,0]
```

```python
#decision tree
classifier = DecisionTreeClassifier()
classifier.fit(xtr,ytr)
```

```python
classifier.score(xtr,ytr)  #Overfitting
```

```python
from sklearn.metrics import accuracy_score
pred = classifier.predict(xtst)
accuracy_score(pred,ytst)
```

```python
#Ensemble

#Random Forest  -  Combination of various Decision tree
rf = RandomForestClassifier(n_estimators=10)
rf.fit(xtr,ytr)
```

```python
rf.score(xtr,ytr)
```

### Bagging

Bagging adalah singkatan dari “bootstrap aggregating”. Ini adalah meta-algoritma, yang mengambil M subsampel (dengan penggantian) dari dataset awal dan melatih model prediktif pada subsampel tersebut. Model akhir diperoleh dengan merata-ratakan model “bootstrapped” dan biasanya menghasilkan hasil yang lebih baik.

Keuntungan utama dari teknik ini adalah menggabungkan regularisasi di dalamnya dan yang Anda butuhkan hanyalah memilih parameter yang baik untuk algoritme dasar. Rata-rata model mengarah pada penghapusan (atau, setidaknya, peningkatan) untuk model yang tidak stabil yang dapat dihasilkan dari data yang bias.

```python
# Bagging

bg = BaggingClassifier(DecisionTreeClassifier(),max_samples=0.5, max_features=1.0, n_estimators=20)
bg.fit(xtr,ytr)
```

```python
bg.score(xtst,ytst)
```
